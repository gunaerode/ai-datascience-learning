```python
# 1. Data collection
# import pandas library - used to work with tables (DataFrames) like excel
import pandas as pd
# reads the csv file and stores it inside the DataFrame called dataset
dataset = pd.read_csv("ml_slr_salary_data.csv")
print("Dataset loaded successfully")
dataset
```

```python
# 2. Data processing
# Select input feature - independent variable
x = dataset[["YearsExperience"]] # Double brackets keep it as a DataFrame, not a series
# Select output/target - dependent variable
y = dataset[["Salary"]]

# Below print statements confirms Input → YearsExperience , Output → Salary
print("\n Independent column:", x.columns)
print("Dependent column:", y.columns)
# Data inspection
# Shows first 5 rows of x and y
print("\nSample X data:")
print(x.head())
print("\nSample y data:")
print(y.head())
```

```python
# 3. Train Test split
# import function to split data into traing and testing data using sklearn
from sklearn.model_selection import train_test_split
# 70% -> Training data
# 30% -> Testing data
# test_size=0.30 -> keeps 30% of data for testing
# random_state=0 -> ensures the same split every time (reproducible)

# Outputs:
# x_train -> training inputs
# x_test  -> testing inputs
# y_train -> training outputs
# y_test  -> testing outputs
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=0)
# Shows number of rows & columns in training input
print("\nX Train shape:", x_train.shape)
print("X Test shape:", x_test.shape)
print("Y Train shape:", y_train.shape)
print("Y Test shape:", y_test.shape)
```

```python
# 4. Model Creation - learning phase 
# Import Linear Regression model
from sklearn.linear_model import LinearRegression
# Model is empty now - no learning yet
regressor = LinearRegression()
# ===== Training phase (Teaching) ===
# Teacher gives questions + answers → student learns
# x_train -> questions
# y_train -> answers
# ========================
# Model learns: Salary = m × YearsExperience + c (m = slope (weight), c = intercept (bias))
# regressor becomes a trained model object and it stores learned parameters + settings
regressor.fit(x_train, y_train) # Model learns y=mX+c using x_train + y_train - It does NOT see test data
print("\nModel trained successfully ✅")

# 5. Model params
weight = regressor.coef_ # slope (m) - salary increase per year of experience
bias = regressor.intercept_ # intercept (c) - salary when experience = 0
print("Weight (slope):", weight)
print("Bias (intercept):", bias)
print("Regression equation: y =", weight, "* X +", bias)

# 6. Predict (test data) - Predict salary for test inputs = x_test -> model -> y_pred (already know x_test -> actual output -> y_test)
# ====== Testing phase (Exam) ====
# Teacher gives new questions -> student answers
# x_test -> new questions
# =========================
y_pred = regressor.predict(x_test)
# Shows predicted salaries
print("\nPredicted values for test data:", y_pred)
```

```python
# 7. Model evalution
# import R² function
# Evaluation phase:
# Student answer vs real answer -> marks
# y_pred -> student answers
# y_test -> real answers
from sklearn.metrics import r2_score
r_score = r2_score(y_test, y_pred)
# Shows model accuracy 1.0 to 0 -> 1 is high accuracy
print("\nR2 Score:", r_score)
# Notes: so far -> The model learns from training data and is evaluated on unseen test data by comparing predicted values with actual values.
```

```python
# 8. Model saving
import pickle
filename = "ml_slr_model.sav" # .sav -> saved model (by convention) You can use any name .pickle
pickle.dump(regressor, open(filename, "wb")) # w -> write, r -> read, b->binary -> Save model "wb"

# You can write the code as shown below instead of using like  pickle.dump(regressor, open(filename, "wb"))
# with open(filename, "wb") as file: # best practice
#    pickle.dump(regressor, file)

print("\nModel saved successfully ✅")

# Load model "rb" w = write, r = read, b = binary

# 9. Model Loading
# Load saved model back into memory
loaded_model = pickle.load(open("ml_slr_model.sav", "rb"))
# with open("finalized_model_linear.sav", "rb") as file:
#    loaded_model = pickle.load(file)

print("Model loaded successfully ✅")
# static year input
# years = 13
# dynamic year
years = int(input("\nEnter Years of Experience: "))
columnName = "YearsExperience"
# Model expects same column name as training data
# So we create a DataFrame with column "YearsExperience"
input_df = pd.DataFrame([[years]], columns=[columnName])

result = loaded_model.predict(input_df)
print("Predicted Salary:", result) # Displays predicted salary
```
