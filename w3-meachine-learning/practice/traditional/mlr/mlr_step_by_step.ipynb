# Multiple Linear Regression (MLR) – Step by Step
This notebook demonstrates how to build a Multiple Linear Regression model using Python and scikit-learn.

Dataset: 50 Startups dataset

Goal: Predict **Profit** based on:
- R&D Spend
- Administration
- Marketing Spend
- State

## Step 1: Import Libraries
We import the necessary Python libraries used for data analysis and machine learning.

```python
# pandas -> used for data manipulation and analysis
import pandas as pd

# numpy -> used for numerical operations
import numpy as np

# pickle -> used to save and load trained machine learning models
import pickle
```

## Step 2: Load the Dataset
We read the dataset using pandas.

```python
# Load dataset from CSV file
dataset = pd.read_csv("mlr_50_startups_data.csv")

# Display first 5 rows
dataset.head()
```

## Step 3: Convert Categorical Data to Numerical Data
Machine learning models cannot understand text values.

The column **State** contains text values like:
- New York
- Florida
- California

We convert them into numbers using **One Hot Encoding**.

```python
# Convert categorical column 'State' into dummy variables
# drop_first=True prevents dummy variable trap
# dtype=int converts True/False into 1/0

dataset = pd.get_dummies(dataset, drop_first=True, dtype=int)

# Check updated dataset
dataset.head()
```

## Step 4: Separate Independent and Dependent Variables

**Independent variables (X)** → Inputs used to predict output

**Dependent variable (y)** → The value we want to predict

Here:
- X = Expenses + State
- y = Profit

```python
# Independent variables (features)
X = dataset[['R&D Spend','Administration','Marketing Spend','State_Florida','State_New York']]

# Dependent variable (target)
y = dataset['Profit']
```

## Step 5: Split Dataset into Training and Testing Data

We split the data into:
- **Training data** → used to train the model
- **Testing data** → used to evaluate the model

Common split:
- 70% training
- 30% testing

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.30,
    random_state=0
)
```

## Step 6: Train the Multiple Linear Regression Model

```python
from sklearn.linear_model import LinearRegression

# Create regression model
regressor = LinearRegression()

# Train model using training data
regressor.fit(X_train, y_train)
```

## Step 7: View Model Coefficients (Weights)

```python
# Coefficients represent how much each feature affects the output
weights = regressor.coef_
print("Weights:", weights)
```

## Step 8: View Intercept (Bias)

```python
# Intercept is the base value when all features are zero
bias = regressor.intercept_
print("Bias:", bias)
```

## Step 9: Predict Test Data

```python
# Predict profit using test dataset
y_pred = regressor.predict(X_test)
```

## Step 10: Evaluate Model Performance

We use **R² Score** to measure model accuracy.

R² value ranges between:
- 0 → poor model
- 1 → perfect model

```python
from sklearn.metrics import r2_score

r_score = r2_score(y_test, y_pred)
print("R2 Score:", r_score)
```

## Step 11: Save the Model

```python
filename = "mlr_practice_model.sav"

# Save trained model
pickle.dump(regressor, open(filename, 'wb'))
```

## Step 12: Load the Model

```python
# Load saved model
loaded_model = pickle.load(open(filename, 'rb'))
```

## Step 13: Predict New Data

```python
# Create new input data with same column structure
new_data = pd.DataFrame(
    [[1234, 345, 4565, 1, 0]],
    columns=['R&D Spend','Administration','Marketing Spend','State_Florida','State_New York']
)

# Predict profit
result = loaded_model.predict(new_data)

print("Predicted Profit:", result)
```
